```python
import csv
import numpy as np
import random
import time

# Create an empty list to store the converted data
converted_data = []
# Open the CSV file and read its contents
with open('hw2.csv', 'r') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    # Iterate through the rows in the CSV file
    for row in csv_reader:
        # Convert each element in the row to a float and append to the converted_data list
        converted_row = [float(element) for element in row]
        converted_data.append(converted_row)
        
# Print the converted data
print(converted_data)
    
# Convert to array
converted_data = np.array(converted_data)

num_cities = len(converted_data)
distance_matrix = np.zeros((num_cities, num_cities))

for i in range(num_cities):
    for j in range(i + 1, num_cities):
        distance = np.linalg.norm(converted_data[i] - converted_data[j])
        distance_matrix[i][j] = distance
        distance_matrix[j][i] = distance
        
# Print the distance matrix
print("Distance Matrix:")
print(distance_matrix)

```

```python
def sim_anneal():
    # Step 3: Implement Evolutionary Algorithm
    # Parameters
    population_size = 100
    mutation_rate = 0.1
    generations = 100

    n = len(distance_matrix)
    city_indices = list(range(n))

    def calculate_tour_cost(solution, distance_matrix):
        num_cities = len(solution)
        tour_cost = 0
        for i in range(num_cities):
            from_city = solution[i]
            to_city = solution[(i + 1) % num_cities]
            tour_cost += distance_matrix[from_city][to_city]
        return tour_cost

    # Create an initial population of solutions
    population = [random.sample(city_indices, n) for _ in range(population_size)]

    # Evaluate the fitness of each solution in the population
    fitness = [1 / calculate_tour_cost(solution, distance_matrix) for solution in population]


    #  Evolutionary Algorithm
    for generation in range(generations):
        # Select parents for the next generation using roulette wheel selection
        parents = random.choices(population, weights=fitness, k=population_size)
        # Create a new generation through mutation and crossover
        new_population = []

        for parent in parents:
            if random.random() < mutation_rate:
                # Apply mutation (swap two random cities)
                mutated_solution = parent[:]
                i, j = random.sample(range(n), 2)
                mutated_solution[i], mutated_solution[j] = mutated_solution[j], mutated_solution[i]
                new_population.append(mutated_solution)
            else:
                # Apply crossover (partially mapped crossover)
                crossover_point = random.randint(0, n - 1)
                parent2 = random.choice(parents)
                child = parent[:crossover_point]
                for city in parent2:
                    if city not in child:
                        child.append(city)
                new_population.append(child)

        population = new_population

        # Evaluate the fitness of the new population
        fitness = [1 / calculate_tour_cost(solution, distance_matrix) for solution in population]

    # Get the best solution from the final generation
    best_solution = population[np.argmax(fitness)]
    best_cost = calculate_tour_cost(best_solution, distance_matrix)
    # Return best solution
    print("Best TSP Tour:", best_solution)
    print("Best TSP Cost:", best_cost)


st = time.time()
sim_anneal()
et = time.time()
print(et-st)
```
